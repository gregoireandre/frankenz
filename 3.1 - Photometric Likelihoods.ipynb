## Photometric Redshifts Overview

At the most basic level, we want to determine how "similar" two objects $g$ and $h$ are based on some associated *observables* $\mathbf{\hat{X}}_g$ and $\mathbf{\hat{X}}_h$, respectively. In other words, given some hypothetical galaxy $i$ with *intrinsic* properties $\mathbf{X}_i$, we want to find out how likely we would observe both $\mathbf{\hat{X}}_g$ and $\mathbf{\hat{X}}_h$. 

Assuming **Poisson independence** between any $g$ and $h$ and marginalizing over $\mathbf{X}_i$, we get

\begin{align}
&\int P(\mathbf{\hat{X}}_g,\mathbf{\hat{X}}_h|\mathbf{X}_i) P(\mathbf{X}_i) \, d\mathbf{X}_i \equiv \int P(g,h|i^\prime) P(i^\prime) \, di^\prime \\ 
&= P(g,h) = P(g|h)P(h) = P(h|g)P(g) \quad . 
\end{align}

where the integral over $di^\prime$ is taken over all possible realizations of $i$.

In practice, $\mathbf{g}$ constitutes a set of **target objects** with **unknown** properties while $\mathbf{h}$ is a set of **training objects** with **known** properties. We are thus interested in $P(h|g)$, which tells us the probability that our observed object $g$ is consistent with our training object $h$. This is just **Bayes theorem**

$$
P(h|g)=\frac{P(g|h)P(h)}{P(g)} 
$$

where $P(h|g)$ is the **posterior** of $h$ given $g$, $P(g|h)$ is the **likelihood** of $g$ given $h$, $P(h)$ is the **prior** for $h$, and

$$
P(g)=\sum_{h \in \mathbf{h}} P(g|h)P(h) 
$$

is the **evidence** for $g$ among all training objects $h \in \mathbf{h}$.

We are interested in inferring the redshift PDF $P(z|g)$ for our observed object $g$. Given $\mathbf{h}$, this is just

$$ 
P(z|g) = \sum_{h\in\mathbf{h}} P(z,h|g) = \sum_{h\in\mathbf{h}} P(z|h) P(h|g) 
= \sum_{h\in\mathbf{h}} P(z|h) \frac{P(g|h)P(h)}{P(g)} \quad , 
$$

which is equivalent to stacking a posterior-weighted set of underlying redshift kernels $P(z|h)$ associated with each training object.

It is important to note that we've made a pretty big assumption here: that we can reduce a continuous process over $\mathbf{X}$ (probed by the **random variable** $\mathbf{\hat{X}}$) to a discrete set of comparisons over our training data $\mathbf{h}$. This choice constitutes a **"Big Data"-oriented approximation** that necessarily introduces some (Poisson) noise into our estimates, and is designed to take advantage of datasets where many ($\gtrsim 10^4$ or so) training objects are available. We will investigate this more explicitly later.

In this particular case, our prior $P(h)$ is defined over a series of models parameterized by magnitude, type, and redshift as outlined in BPZ (Benitez 2000). See Part 1 for more details.

```python
# define magnitude prior
def pmag(m, mlim): # magnitude grid and magnitude 5-sigma limit
    alpha, beta, off = 15.0, 2., 1. # coefficients
    p_imag = m**alpha*exp(-(m/(mlim-off))**beta)
    p_imag /= p_imag.sum()
    
    return p_imag

# define the original BPZ prior
def bpz_prior(m, zgrid): # m = magnitude (~8140A), z = redshift
        
    # Formula: zm = zo + km*dm,  p(z|T,m) = z**a * exp(-(z/zm)**a)
    # coefficients from Table 1 of Benitez00
    a  = array([2.465,  1.806,  0.906])
    zo = array([0.431,  0.390,  0.0626])
    km = array([0.0913, 0.0636, 0.123])
    k_t= array([0.450,  0.147])

    # Fractions expected at m=20: 35% E/S0, 50% Spiral, 15% Irr
    fo_t = array([0.35, 0.5])

    # establish magnitude bounds
    mmin, mmax = 20., 32.
    m=clip(m, mmin, mmax) 
    dm = m - mmin # dmag
    
    # establish redshift bounds
    zmin, zmax = 0., 15.
    zmt = clip(zo+km*dm, zmin, zmax)
    zmt_at_a = zmt**a
    zt_at_a = power.outer(zgrid, a)
    
    # morphological fractions (0=Ell/S0, 1=Spiral, 2=Irr)
    f_t = zeros(3)
    f_t[:2] = fo_t * np.exp(-k_t*dm)
    f_t[2] = 1 - f_t.sum()
    
    # compute probability
    p_i = zt_at_a * exp(-clip(zt_at_a/zmt_at_a,0.,700.))
    p_i /= p_i.sum(axis=0)
    p_i *= f_t
    
    return p_i, f_t

# compute results over an (m,z) grid
mgrid = linspace(20., 32., 1000) # magnitude grid
zgrid = linspace(0., 15., 1000) # redshift grid

# define our N-D linteraly interpolated priors
bpz_arr = array([bpz_prior(m, zgrid)[0] for m in mgrid]) # priors at (m,z,t)
prior1 = interpolate.RegularGridInterpolator((mgrid, zgrid, [0,1,2]), bpz_arr) # P(t,z|m)
bpz_arr = array([bpz_prior(m, zgrid)[1] for m in mgrid]) # priors at (m,t)
prior2 = interpolate.RegularGridInterpolator((mgrid, [0,1,2]), bpz_arr) # P(t|m)

# define P(t,z|m)
def p_tz_m(t, z, m):
    m, z = clip(m, 20., 32.), clip(z, 0., 15.)
    return prior1((m, z, t))

# define P(t|m)
def p_t_m(t, m):
    m = clip(m, 20., 32.)
    return prior2((m, t))
```

## Photometric Likelihoods

Our observables for $g$ and $h$ can be modeled as **kernel density estimates (KDEs)** $K_g(\mathbf{F}|\mathbf{\hat{F}}_g)$ and $K_h(\mathbf{F}|\mathbf{\hat{F}}_h)$, where $\mathbf{\hat{F}}_g = \lbrace \hat{F}_{g,1},\ldots,\hat{F}_{g,n} \rbrace$ and $\mathbf{\hat{F}}_h=\lbrace \hat{F}_{h,1},\ldots,\hat{F}_{h,n} \rbrace$ are the corresponding set of $n$ observables associated with galaxies $h$ and $g$, respectively. These are most often a set of observed **flux densities** within a set of bands $\mathbf{b}$ (as suggested by the notation), but can also include additional observables such as a galaxy's shape or size.

Our likelihood is then
$$ 
P(g|h) = \int P(\mathbf{F}) P(g|h,\mathbf{F}) \, d\mathbf{F} 
= \int P(\mathbf{F}) K_g(\mathbf{F}|\mathbf{\hat{F}}_g) \,  K_h(\mathbf{F}|\mathbf{\hat{F}}_h) \, d\mathbf{F} 
$$ 

where $P(\mathbf{F})$ is a prior over the flux densities across $\mathbf{b}$.

For most galaxies, we can approximate $K_g(\mathbf{F}|\mathbf{\hat{F}}_h)$ as a multivariate Normal (i.e. Gaussian) distribution such that

$$ 
K_g(\mathbf{F}|\mathbf{\hat{F}}_g) = \mathcal{N}(\mathbf{F}|\mathbf{\hat{F}}_g,\mathbf{\hat{C}}_g) 
\equiv \frac{\exp\left[-\frac{1}{2}||\mathbf{F}-\mathbf{\hat{F}}||_{\mathbf{\hat{C}}^{-1}}\right]}{(2\pi)^{N_b/2}||\mathbf{\hat{C}}_g||^{1/2}} 
$$

where

$$ 
||\mathbf{F}-\mathbf{\hat{F}}||_{\mathbf{\hat{C}}^{-1}} \equiv (\mathbf{F}-\mathbf{\hat{F}}_g)^T\mathbf{\hat{C}}_g^{-1}(\mathbf{F}-\mathbf{\hat{F}}_g) 
$$

is the squared Mahalanobis distance or **Mahalanobis norm** for vector $\mathbf{F}-\mathbf{\hat{F}}$, $\mathbf{\hat{C}}_g$ is the estimated **covariance matrix**, $\mathbf{\hat{C}}_g^{-1}$ is the corresponding **precision matrix**, $T$ is the transpose operator, $p$ is the total number of bands, and

$$
||\mathbf{\hat{C}}|| \equiv ||\mathrm{vec}(\mathbf{\hat{C}})|| 
= \mathrm{vec}(\mathbf{\hat{C}})^T \mathrm{vec}(\mathbf{\hat{C}}) 
= \sum_{i,j} \hat{C}_{ij}^2
$$

is the **Euclidean matrix norm** where the $\mathrm{vec}(\cdot)$ operator "vectorizes" an $n \times m$ matrix by stringing it out into an $nm \times 1$ column vector.

### Note that while all of our formal derivations will use matrix notation, in practice we assume all our covariances are diagonal (i.e. the errors are independent).

## Likelihood: Flux Densities

We first look at the simplest case: a direct observational comparison over $\mathbf{F}$ (i.e. galaxy **magnitudes**).

The product of two multivariate Normal distributions $\mathcal{N}(\mathbf{F}|\mathbf{\hat{F}}_g,\mathbf{\hat{C}}_g)$ and $\mathcal{N}(\mathbf{F}|\mathbf{\hat{F}}_h,\mathbf{\hat{C}}_h)$ is a scaled multivariate Normal of the form $S_{gh}\,\mathcal{N}(\mathbf{F}|\mathbf{\hat{F}}_{gh},\mathbf{\hat{C}}_{gh})$ where

$$
S_{gh} \equiv \mathcal{N}(\Delta\mathbf{\hat{F}}_{gh}|\mathbf{0},\mathbf{\hat{C}}_{g+h}), \quad 
\Delta\mathbf{\hat{F}}_{gh} \equiv \mathbf{\hat{F}}_{g}-\mathbf{\hat{F}}_{h}, \quad
\mathbf{\hat{C}}_{g+h} \equiv \mathbf{\hat{C}}_g+\mathbf{\hat{C}}_h, \quad
\mathbf{\hat{F}}_{gh} \equiv \mathbf{\hat{C}}_{gh} \left( \mathbf{\hat{C}}_g^{-1}\mathbf{\hat{F}}_g + \mathbf{\hat{C}}_h^{-1}\mathbf{\hat{F}}_h \right), \quad
\mathbf{\hat{C}}_{gh} \equiv \left(\mathbf{\hat{C}}_g^{-1}+\mathbf{\hat{C}}_h^{-1}\right)^{-1} \quad .
$$

If we assume a uniform prior on our flux densities and marginalize over $\mathbf{F}$, our log-likelihood can then be written as

\begin{equation}
\boxed{
-2\ln P(g|h) = ||\Delta\mathbf{\hat{F}}_{gh}||_{\mathbf{\hat{C}}_{g+h}^{-1}} + \delta(g|h)
}
\end{equation}

where $\delta(g|h)$ is some constant as a function of $g$ and $h$ that we will subsequently ignore (for reasons that will be explained in Notebook 3).

Let's compute an example PDF using **FRANKEN-Z** for objects in our mock catalog. Since these are sampled from the prior, we've actually introduced our prior *implicitly* via the *distribution of objects* in our training sample. As a result, computing likelihoods directly in magnitudes actually probes the full *posterior* (as defined by BPZ). 

We compare two versions of our results: 
- **Noiseless case**: computed using the "true" underlying photometry underlying each training object. By definition, this should probe the "true" posterior.
- **Noisy case**: computed using our "observed" mock photometry. By construction, this should be a noisy but unbiased probe of the "true" posterior.

## Likelihood: Colors

We can also defined our likelihoods in terms of **flux ratios** (i.e. galaxy **colors**) by introducing a scaling parameter $\alpha$. This takes the form

$$
P_C(g|h) = \int P(\alpha) \mathcal{N}\left(\Delta\mathbf{\hat{F}}_{gh}(\alpha) \, \middle| \, \mathbf{0},\mathbf{\hat{C}}_{g+h}(\alpha)\right)\,d\alpha
$$

where 

$$
\Delta \mathbf{\hat{F}}_{gh}(\alpha) \equiv \mathbf{\hat{F}}_{g} - \alpha\mathbf{\hat{F}}_{h}, \quad
\mathbf{\hat{C}}_{g+h}(\alpha) \equiv \mathbf{\hat{C}}_{g}+\alpha^2\mathbf{\hat{C}}_{h}
$$

and we have assumed $P(\mathbf{F},\alpha)=P(\mathbf{F})P(\alpha)$ and that $P(\mathbf{F})$ is uniform.

This integral does not have an analytic solution (but converges under a uniform prior over $\alpha$; see Speagle et al. 2017 for more details). However, **if we assume that $\mathbf{\hat{C}}_h = \mathbf{0}$**, then there is an analytic solution with log-likelihood

\begin{equation}
\boxed{
-2 \ln P_C(g|h) = ||\Delta\mathbf{\hat{F}}_{gh}(\alpha_{\mathrm{ML}})||_{\mathbf{\hat{C}}_g^{-1}} + \delta_C(g|h)
}
\end{equation}

where

$$
\alpha_{\mathrm{ML}} = \frac{\gamma^2}{2}\left(\mathbf{\hat{F}}_g^T\mathbf{\hat{C}}_g^{-1}\mathbf{\hat{F}}_h 
+ \mathbf{\hat{F}}_h^T \mathbf{\hat{C}}_g^{-1} \mathbf{\hat{F}}_g\right), \quad
\gamma = \left(\mathbf{\hat{F}}_h^T\mathbf{\hat{C}}_g^{-1}\mathbf{\hat{F}}_h^T\right)^{-1/2}
$$

and $\delta_C(g|h)$ is again some constant.

We can now repeat the above exercise using our color-based likelihoods. We compare two versions:
- **Likelihood case**: using our pre-computed grid of models to compute color-based likelihoods only.
- **Posterior case**: using the BPZ prior to compute the actual posterior. By construction, this should be equivalent to the true posterior.

```python
# read labels
Nt, Nz = len(sdss.models_t), len(sdss.models_z)
mphot = sdss.models_phot.reshape(Nt*Nz,sdss.NFILTER) # model photometry

# compute log-likelihoods
ll, nbands, s = fz.loglikelihood_s(sdss.phot_obs[idx], sdss.noise[idx], ones(sdss.NFILTER), # target object
                                   mphot, zeros(mphot.shape), ones(mphot.shape), 
                                   return_s=True) # training data
like = exp(-0.5*ll) # convert to likelihood
like_max = max(like) # maximum-likelihood solution (will be the same object here)

# compute likelihoods
like = exp(-0.5*ll) # convert to likelihood
like_max = like.max() # maximum-likelihood solution (will be the same object here)
sel=arange(len(mphot))[like > 1e-2*like_max] # select obly higher-likelihood objects

# compute posteriors
prior = array([p_tz_m(t, sdss.models_z, mags[idx]) for t in sdss.types]).flatten()
post = prior * like
post_max = post.max()

# plot flux distribution
figure(figsize=(16,14))
for i in choice(sel,size=500):
    subplot(3,2,1)
    plot(sdss.lambda_eff, s[i]*mphot[i], alpha=like[i]/like_max*0.1, lw=3, color='blue', zorder=1)
    subplot(3,2,2)
    plot(sdss.lambda_eff, s[i]*mphot[i], alpha=post[i]/post_max*0.1, lw=3, color='red', zorder=1)
subplot(3,2,1)
errorbar(sdss.lambda_eff, sdss.phot_obs[idx], yerr=sdss.noise[idx], lw=3, color='black', zorder=2)
xlabel(r'Wavelength ($\AA$)')
xlim([sdss.lambda_eff.min()-100, sdss.lambda_eff.max()+100])
ylim([(sdss.phot_obs[idx]-sdss.noise[idx]).min()*0.9, (sdss.phot_obs[idx]+sdss.noise[idx]).max()*1.1])
ylabel(r'$F_\nu$')
yticks(fontsize=24)
title('Likelihood', y=1.04)
tight_layout()
subplot(3,2,2)
errorbar(sdss.lambda_eff, sdss.phot_obs[idx], yerr=sdss.noise[idx], lw=3, color='black', zorder=2)
xlabel(r'Wavelength ($\AA$)')
xlim([sdss.lambda_eff.min()-100, sdss.lambda_eff.max()+100])
ylim([(sdss.phot_obs[idx]-sdss.noise[idx]).min()*0.9, (sdss.phot_obs[idx]+sdss.noise[idx]).max()*1.1])
ylabel(r'$F_\nu$')
yticks(fontsize=24)
title('Posterior', y=1.04)
tight_layout()

t, z = meshgrid(sdss.models_t, sdss.models_z) # compute coordinates
t, z = t.swapaxes(0,1).flatten(), z.swapaxes(0,1).flatten() # swap to proper order
tgrid, zgrid = arange(sdss.NTEMPLATES+1)-0.5, arange(0,6.01,0.05) # define 2-D bins

# plot redshift PDF
subplot(3,2,3)
h = hist(z, bins=zgrid, weights=like, histtype='stepfilled', lw=2, color='yellow')
vlines(redshifts[idx], 0., h[0].max(), color='red', linestyles='--', lw=2)
xlabel('Redshift')
ylabel('PDF')
ylim([0., h[0].max()])
yticks([])
title('Likelihood', y=1.02)
tight_layout()
subplot(3,2,4)
h = hist(z, bins=zgrid, weights=post, histtype='stepfilled', lw=2, color='yellow')
vlines(redshifts[idx], 0., h[0].max(), color='red', linestyles='--', lw=2)
xlabel('Redshift')
ylabel('PDF')
ylim([0., h[0].max()])
yticks([])
title('Posterior', y=1.02)
tight_layout()

# plot redshift-template joint distribution
subplot(3,2,5)
hh = hist2d(z, t, bins=[zgrid, tgrid], weights=like/like.sum(), cmin=2e-4, norm=mpl.colors.LogNorm())
hlines(templates[idx], zgrid.min(), zgrid.max(), color='black', lw=2)
vlines(redshifts[idx], tgrid.min(), tgrid.max(), color='black', lw=2)
colorbar()
xlabel('Redshift')
ylabel('Template')
title('Likelihood', y=1.04)
tight_layout()
subplot(3,2,6)
hh = hist2d(z, t, bins=[zgrid, tgrid], weights=post/post.sum(), cmin=2e-4, norm=mpl.colors.LogNorm())
hlines(templates[idx], zgrid.min(), zgrid.max(), color='black', lw=2)
vlines(redshifts[idx], tgrid.min(), tgrid.max(), color='black', lw=2)
colorbar()
xlabel('Redshift')
ylabel('Template')
title('Posterior', y=1.04)
tight_layout()
```

As expected, the secondary solutions seen in our grid-based likelihoods are suppressed by our prior, which indicates many of these solutions are distinctly unphysical (at least given the original assumptions used when constructing our mock).

### Sidenote: KDE and FRANKEN-Z

One of the ways FRANKEN-Z differs from other photometric redshift (photo-z) codes is that it tries to avoid discretizing quantities whenever and wherever possible. Since redshifts, flux densities, and many other photometric quantities are **continuous with smooth PDFs**, we attempt to work directly in this continuous space whenever possible instead of resorting to binning.

As in the last notebook, we accomplish this through **kernel density estimation (KDE)**. Since almost all photometric observable PDFs are Gaussian, by connecting each observable with an associated Gaussian kernel density we can *in theory* construct a density estimate at any location in parameter space by evaluating the probability density of all kernels at that location.

In practice, such a brute-force approach is prohibitively computationally expensive. Instead, we approximate the contribution from any particular object by (1) evaluating only a small subset of "nearby" kernels, (2) evaluating the overall kernel density estimates over a discrete basis, and (3) evaluating only the "central regions" of our kernels. This is implemented through the *pdf_kde(...)* function in FRANKEN-Z.

In addition, we can also use a stationary **pre-computed Gaussian kernel dictionary** to discretize our operations. This avoids repetitive, expensive computations at the (very small) cost of increasing overhead and resolution penalties. This is implemented via *pdf_kde_dict(...)* and is what we'll generally be using.

Our dictionary parameters are stored in files referenced from a master configuration file. Let's initialize these now.

```python
# initialize asinh magnitudes (Luptitudes)
flux_zeropoint = 10**(-0.4*-23.9) # AB magnitude zeropoint
mag, magerr = fz.asinh_mag_map(sdss.phot_obs, sdss.noise, # map f_nu to asinh mag
                               skynoise=sdss.FDEPTHS, zeropoints=flux_zeropoint) # softening parameter and zero-point

# read FRANKEN-Z config files (to get dictionaries)
config = fz.ReadParams('config/frankenz.config')

# initialize supplementary dictionaries
mdict = fz.PDFDict(config.CONFIG_MAG.params) # magnitude dict
cdict = fz.PDFDict(config.CONFIG_COLOR.params) # color dict
magidx, mageidx = mdict.fit(mag, magerr) # discretized magnitudes
```

Now that we've successfully initialized our magnitude dictionary, let's see what the magnitude distribution of our catalog looks like.

```python
# plotting magnitude distribution
fcolors=get_cmap('winter')(linspace(0,1,sdss.NFILTER)) # filter colors
figure(figsize=(14,6))
for i in xrange(sdss.NFILTER):
    subplot(2,sdss.NFILTER,i+1)
    mag_pdf = fz.pdf_kde_dict(mdict.sig_dict, mdict.sig_width, magidx[:,i], mageidx[:,i], sel.astype(int),
                              mdict.grid, mdict.delta, Nobs, mdict.Ngrid) # smooth PDF generated using input dictionaries
    plot(mdict.grid, mag_pdf, lw=3, color=fcolors[i])
    vlines(sdss.MDEPTHS[i], mag_pdf.max()*1.05, mag_pdf.min(), color=fcolors[i], lw=3, linestyles='--')
    xlim([16, 30])
    yticks([])
    ylim([None,mag_pdf.max()*1.05])
    xticks(arange(16,30,4), fontsize=18)
    xlabel(sdss.filters[i], fontsize=20)
    tight_layout()
    subplot(2,sdss.NFILTER,i+1+sdss.NFILTER)
    h, bins = histogram(magerr[:,i][sel], normed=True, range=[0,0.6], bins=100)
    plot((bins[:-1]+bins[1:])/2.,h,lw=3,color=fcolors[i])
    xticks(linspace(0,0.6,4), fontsize=18)
    xlabel(sdss.filters[i]+'_err', fontsize=20)
    yticks([])
    tight_layout()
```

Note that we've used asinh magnitudes (i.e. "Luptitudes"; [Lupton et al. 1999](http://adsabs.harvard.edu/abs/1999AJ....118.1406L)) rather than log10 magnitudes in order to incorporate data with negative measured fluxes.

It is important to note that our assumption about translational-invariant kernels **does not** hold in redshift space, which is multiplicative (and generally has scale-dependent errors). However, it **does** hold in $\ln(1+z)$ space (which is probably a more natural place to do fitting anyway). FRANKEN-Z thus natively computes redshift PDFs in $\ln(1+z)$ space, although quantities can easily be converted back back to the more typically used redshift-space PDFs.

```python
# initialize redshift dictionary
rdict = fz.RedshiftDict(config.CONFIG_REDSHIFT.params)

# add names so easier to reference later
rdict.sig_dict = rdict.lze_dict
rdict.sig_width = rdict.lze_width
rdict.delta = rdict.dlz_highres
rdict.grid = rdict.lzgrid_highres
rdict.Ngrid = rdict.Nz_highres

# discretize redshifts
lzidx, lzeidx = rdict.fit(log(1+redshifts), ones(Nobs)*rdict.dlz) # discretized ln(1+z)
```

Let us now compute the redshift PDF.

**Important note**: When calculating our PDFs in $\ln(1+z)$ space, we actually compute them at a higher-resolution than required so that any downsampled, interpolated PDFs will be more accurate. We also include left-/right-padding to avoid edge effects. Also, when recasting our PDFs from $\ln(1+z)$ to redshift space, we need to correct for the change from logarithmic to linear density, i.e. $d \ln(1+z) = dz/(1+z)$.

```python
# plotting redshift distribution
figure(figsize=(14,6))

# all
z_pdf = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx, lzeidx, ones(Nobs),
                       rdict.grid, rdict.delta, Nobs, rdict.Ngrid) # smooth PDF generated using input dictionaries
z_pdf = z_pdf[rdict.zmin_idx_highres:rdict.zmax_idx_highres] / (1+rdict.zgrid_highres) # ln(1+z)->z density
z_pdf /= z_pdf.sum() # normalizing
plot(rdict.zgrid_highres, z_pdf, lw=5, color='black')
fill_between(rdict.zgrid_highres, z_pdf, color='gray', alpha=0.4, label='Underlying')

# observed
z_pdf = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx, lzeidx, sel.astype(int),
                        rdict.grid, rdict.delta, Nobs, rdict.Ngrid) # smooth PDF generated using input dictionaries
z_pdf = z_pdf[rdict.zmin_idx_highres:rdict.zmax_idx_highres] / (1+rdict.zgrid_highres) # ln(1+z)->z density
z_pdf /= z_pdf.sum() # normalizing
plot(rdict.zgrid_highres, z_pdf, lw=5, color='navy')
fill_between(rdict.zgrid_highres, z_pdf, color='blue', alpha=0.4, label='Observed')
xlim([0, 6])
yticks([])
legend(fontsize=20)
tight_layout()
xlabel('Redshift')
ylabel('$N(z|\mathbf{g})$')
```

## Comparison 3: Brute-Force vs Sparse

We've now demonstrated that:
1. The original BPZ posterior is an unbiased probe of the underlying distribution (as expected).
2. Our magnitude-based likelihoods computed over unbiased training data provides is also an unbiased (but noisier) probe of the underlying distribution (as expected).

We now turn to the computational challenge of exploiting this training data. In particular, taking a brute-force approach (as we did above) is prohibitively slow. It also penalizes you harshly for expanding your training sample, which seems counter-productive.

To solve this issue (somewhat foreshadowed in our KDE discussion above), we turn to machine learning. Specifically, we want to use machine learning to give us a **sparse approximation of the true likelihood distribution using a small but unbiased set of neighbors**. In other words, we want to select a small subset of objects ($N_{\mathrm{neighbor}} \ll N_{\mathrm{train}}$) and use the likelihoods evaluated over only those objects as an approximation of the overall likelihood.

### Sidenote: Nearest Neighbors and FRANKEN-Z

We use **k-d trees** to find neighbors in $\mathcal{O}(\log N)$ time (although this does strongly depend on dimensionality), which is a much softer penalty than the $\mathcal{O}(N)$ scaling imposed by brute-force approaches. Using an ensemble of k-d trees trained/queried with Monte Carlo realizations of the input fluxes, we can subsequently guarantee (in the limit of many objects and/or draws) an unbiased selection of (unique) neighbors gathered from across the ensemble that properly takes into account measurement errors within both datasets.

Once we have this sparse representation of the likelihoods in hand, we can then map them onto **any** target distribution using weighted KDE (as we've done above).

```python
# initialize FRANKEN-Z instance
frankenz=fz.FRANKENZ(N_members=25) # 25 trees
```

Once we've initialized our FRANKEN-Z instance, predictions can be generated extremely quickly using the associated *predict(...)* function.

```python
# generate unique model predictions and corresponding log-likelihoods
model_obj, model_Nobj, model_ll, model_Nbands = frankenz.predict(phot_train, err_train, mask_train, 
                                                                 phot_test, err_test, mask_test)
```

Now let's turn these predictions into the population estimates we were previously using.

```python
pstack_fz = zeros(Ntrain)
for i in xrange(Ntest):
    Nm = model_Nobj[i]
    midx = model_obj[i][:Nm]
    ll = model_ll[i][:Nm]
    like = exp(-0.5*(ll-ll.min()))
    pstack_fz[midx] += like/like.sum()
    
# computing redshift distribution (sparse)
z_pdf_fz = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx[train_sel], lzeidx[train_sel], pstack_fz, 
                           rdict.grid, rdict.delta, Ntrain, rdict.Ngrid)
z_pdf_fz = z_pdf_fz[rdict.zmin_idx_highres:rdict.zmax_idx_highres] / (1+rdict.zgrid_highres)
z_pdf_fz /= z_pdf_fz.sum()
```

```python
# plotting redshift distribution
figure(figsize=(14,6))
plot(rdict.zgrid_highres, zpdf, lw=5, color='black', label='Underlying')
plot(rdict.zgrid_highres, z_pdf, lw=5, color='red', alpha=0.7, label='Mag Likelihood (brute-force)')
plot(rdict.zgrid_highres, z_pdf_fz, lw=5, color='darkviolet', alpha=0.7, label='Mag Likelihood (sparse)')
xlim([z.min(), z.max()])
yticks([])
legend(fontsize=20)
tight_layout()
xlabel('Redshift')
ylabel('$N(z|\mathbf{g})$')

# plot redshift-template joint distribution
figure(figsize=(16,5))
subplot(1,2,1)
hh = hist2d(redshifts[train_sel], templates[train_sel], bins=[zgrid, tgrid], 
            weights=pstack1/pstack1.sum(), cmin=2e-4, norm=mpl.colors.LogNorm())
colorbar()
xlabel('Redshift')
ylabel('Template')
title('Mag Likelihood (brute-force)', y=1.04, fontsize=26)
tight_layout()
subplot(1,2,2)
hh = hist2d(redshifts[train_sel], templates[train_sel], bins=[zgrid, tgrid], 
            weights=pstack_fz/pstack_fz.sum(), cmin=2e-4, norm=mpl.colors.LogNorm())
colorbar()
xlabel('Redshift')
ylabel('Template')
title('Mag Likelihood (sparse)', y=1.04, fontsize=26)
tight_layout()
```

These distributions are nearly identical, illustrating that our sparse, nearest-neighbor approximation works well.

## Comparison 4: Dependence on Neighbors and Sample Size

Due to the way we utilize training data (treating all objects as discrete comparisons rather than as samples from a continuous distribution), our comparisons using training data will be subject to some amount of Poisson (i.e. shot) noise based on our overall sample size and our effective number of neighbors. We want to quantify this below.

```python
Ntrain, Ntest = 100000, 50000
train_sel = arange(sdss.NSAMPLES)[sel][:Ntrain] # training set
test_sel = arange(sdss.NSAMPLES)[sel][Ntrain:Ntrain+Ntest] # testing set

phot_train, phot_test = sdss.phot_obs[train_sel], sdss.phot_obs[test_sel]
err_train, err_test = sdss.noise[train_sel], sdss.noise[test_sel]
mask_train, mask_test = ones(phot_train.shape), ones(phot_test.shape)
```

### Neighbor-dependence

Here we keep the overall number of objects fixed at 100k (large sample), but vary the number of trees used in our approximation. Since each tree probes a separate Monte Carlo realization of our observables, this probes both the scope of our search as well as our ability to incorporate measurement errors.

```python
# change trees
ntrees = [1, 2, 4, 6, 10, 15, 25, 40, 60, 100]

pstack_fz, z_pdf_fz = [], []

for n in ntrees:
    frankenz = fz.FRANKENZ(N_members=n) # 25 trees
    sys.stdout.write(str(n)+': ')

    # compute predictions
    model_obj, model_Nobj, model_ll, model_Nbands = frankenz.predict(phot_train, err_train, mask_train, 
                                                                     phot_test, err_test, mask_test)

    # compute evidence
    pstack_fz_t = zeros(Ntrain)
    for i in xrange(Ntest):
        Nm = model_Nobj[i]
        midx = model_obj[i][:Nm]
        ll = model_ll[i][:Nm]
        like = exp(-0.5*(ll-ll.min()))
        pstack_fz_t[midx] += like/like.sum()

    # computing redshift distribution (sparse)
    z_pdf_fz_t = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx[train_sel], lzeidx[train_sel], 
                                 pstack_fz_t, rdict.grid, rdict.delta, Ntrain, rdict.Ngrid)
    z_pdf_fz_t = z_pdf_fz_t[rdict.zmin_idx_highres:rdict.zmax_idx_highres] / (1+rdict.zgrid_highres)
    z_pdf_fz_t /= z_pdf_fz_t.sum() # normalizing

    pstack_fz.append(pstack_fz_t.copy())
    z_pdf_fz.append(z_pdf_fz_t.copy())
```

```python
# truth
zpdf = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx[test_sel], lzeidx[test_sel], ones(Ntest),
                       rdict.grid, rdict.delta, Ntest, rdict.Ngrid)
zpdf = zpdf[rdict.zmin_idx_highres:rdict.zmax_idx_highres] / (1+rdict.zgrid_highres)
zpdf /= zpdf.sum() # normalizing

colors=get_cmap('rainbow')(linspace(0,1,len(ntrees)))

# plotting
figure(figsize=(14,6))
plot(rdict.zgrid_highres, zpdf, lw=5, color='black', label='Truth')
for i in xrange(len(pstack_fz)):
    plot(rdict.zgrid_highres, z_pdf_fz[i], lw=3, color=colors[i], alpha=0.5, label='Ntree='+str(ntrees[i]))
xlim([z.min(), z.max()])
yticks([])
legend(fontsize=20, ncol=3)
tight_layout()
xlabel('Redshift')
ylabel('$N(z|\mathbf{g})$')

# plotting Poisson deviations
Sp = [fz.compute_density_score(zpdf*Ntest, z_pdf_fz[i]*Ntest)[0][0] for i in xrange(len(pstack_fz))] # "Poisson deviation"
figure(figsize=(14,6))
for i in xrange(len(pstack_fz)):
    semilogx(ntrees[i], Sp[i]/len(zpdf), 'o', markersize=20, color=colors[i], alpha=0.5)
xlim([min(ntrees)*0.9, max(ntrees)*1.1])
ylim([(min(Sp)-5)/len(zpdf), (max(Sp)+5)/len(zpdf)])
tight_layout()
xlabel('N_trees')
ylabel('Normalized Deviation', fontsize=28)
```

In the limit where we have plenty of training data (i.e. good coverage), our population results appear relatively robust to the number of trees (and hence neighbors) when aggregated across our ensemble. Put another way, the variance in our estimates is essentially negligible. 

**Note: This ignores the impact of additional trees on predictions for individual objects, where it's obvious including more neighbors will improve the corresponding PDFs.**

### Sample-dependence

Now we fix the number of trees to 25 and vary our original sample size.

```python
# initialize
frankenz = fz.FRANKENZ(N_members=25)

# downsample
downsample = [100, 75, 50, 35, 20, 10, 5, 2, 1]

pstack_fz, z_pdf_fz = [], []

for down in downsample:
    sys.stdout.write(str(down)+': ')
    phot_train, phot_test = sdss.phot_obs[train_sel][::down], sdss.phot_obs[test_sel]
    err_train, err_test = sdss.noise[train_sel][::down], sdss.noise[test_sel]
    mask_train, mask_test = ones(phot_train.shape), ones(phot_test.shape)
    Ndown = len(phot_train)

    # compute predictions
    model_obj, model_Nobj, model_ll, model_Nbands = frankenz.predict(phot_train, err_train, mask_train, 
                                                                     phot_test, err_test, mask_test)

    # compute evidence
    pstack_fz_t = zeros(Ndown)
    for i in xrange(Ntest):
        Nm = model_Nobj[i]
        midx = model_obj[i][:Nm]
        ll = model_ll[i][:Nm]
        like = exp(-0.5*(ll-ll.min()))
        pstack_fz_t[midx] += like/like.sum()

    # computing dN/dz (sparse)
    z_pdf_fz_t = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx[train_sel][::down], lzeidx[train_sel][::down], 
                                   pstack_fz_t, rdict.grid, rdict.delta, Ndown, rdict.Ngrid)
    z_pdf_fz_t = z_pdf_fz_t[rdict.zmin_idx_highres:rdict.zmax_idx_highres] / (1+rdict.zgrid_highres) # ln(1+z)->z density
    z_pdf_fz_t /= z_pdf_fz_t.sum() # normalizing

    pstack_fz.append(pstack_fz_t.copy())
    z_pdf_fz.append(z_pdf_fz_t.copy())
```

```python
colors = get_cmap('rainbow')(linspace(0,1,len(ntrees)))

# plotting dN/dz
figure(figsize=(14,6))
plot(rdict.zgrid_highres, zpdf, lw=5, color='black', label='Truth')
for i in xrange(len(pstack_fz)):
    Ndown = len(train_sel[::downsample[i]])
    plot(rdict.zgrid_highres, z_pdf_fz[i], lw=3, color=colors[i], alpha=0.5, label='N='+str(Ndown))
xlim([z.min(), z.max()])
yticks([])
legend(fontsize=20, ncol=3)
tight_layout()
xlabel('Redshift')
ylabel('$N(z|\mathbf{g})$')

# plotting MSE
Ndown = [len(train_sel[::downsample[i]]) for i in xrange(len(downsample))]
Sp = [fz.compute_density_score(zpdf*Ntest, z_pdf_fz[i]*Ntest)[0][0] for i in xrange(len(pstack_fz))] # "Poisson deviation"
figure(figsize=(14,6))
for i in xrange(len(pstack_fz)):
    semilogx(Ndown[i], Sp[i]/len(zpdf), 'o', markersize=20, color=colors[i], alpha=0.5)
xlim([min(Ndown)*0.9, max(Ndown)*1.1])
ylim([(min(Sp)-100)/len(zpdf), (max(Sp)+100)/len(zpdf)])
tight_layout()
xlabel('N_train')
ylabel('Normalized Deviation', fontsize=28)
```

We see that for our mock SDSS data, our gains appear to top out around 20-50k objects. We also note that the improvement here is pretty significant compared with the number of trees we utilize to compute our estimates. In other words, for most applications, getting more training objects (especially in regimes with poor sampling) will likely bring about significantly greater gains than improving the number of neighbors used in our fits (at least when it comes to our population estimates).

## Quantifying Performance

To wrap things up, we want quantify how well we do here with our "fiducial" model, where train on 50% of the data using 25 trees.

```python
# initialize samples
Ntrain, Ntest = int(Nsel/2.), Nsel-int(Nsel/2.)
train_sel = arange(sdss.NSAMPLES)[sel][:Ntrain] # training set
test_sel = arange(sdss.NSAMPLES)[sel][Ntrain:Ntrain+Ntest] # testing set

phot_train, phot_test = sdss.phot_obs[train_sel], sdss.phot_obs[test_sel]
err_train, err_test = sdss.noise[train_sel], sdss.noise[test_sel]
mask_train, mask_test = ones(phot_train.shape), ones(phot_test.shape)
```

```python
# initialize FRANKEN-Z instance
frankenz=fz.FRANKENZ(N_members=25) # 25 trees

# generate predictions
model_obj, model_Nobj, model_ll, model_Nbands = frankenz.predict(phot_train, err_train, mask_train, 
                                                                 phot_test, err_test, mask_test)
```

We now convert our original likelihoods into individual redshift PDFs.

```python
# generate redshift predictions
model_zpdf = zeros((Ntest,len(rdict.zgrid)))
model_pstack = zeros(Ntrain)

for i in xrange(Ntest):
    if i%5000 == 0: sys.stdout.write(str(i)+' ')
    Nm = model_Nobj[i]
    midx = model_obj[i][:Nm]
    ll = model_ll[i][:Nm]
    like = exp(-0.5*(ll-ll.min()))
    prob = like/like.sum()
    model_pstack[midx] += prob

    pdf = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx[train_sel[midx]], lzeidx[train_sel[midx]], 
                          prob, rdict.grid, rdict.delta, Nm, rdict.Ngrid)
    pdf = pdf[rdict.zmin_idx_highres:rdict.zmax_idx_highres:rdict.res] / (1+rdict.zgrid) # downsample, convert from ln(1+z)->z
    model_zpdf[i] = pdf

model_zpdf = fz.pdfs_resample(rdict.zgrid, model_zpdf, rdict.zgrid_out)
```

### Population Comparison

Let's see how well the population redshift distribution $N(z|\mathbf{g})$ distribution computed actually matches the underlying distribution.

```python
# truth
zpdf = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx[test_sel], lzeidx[test_sel], ones(Ntest),
                       rdict.grid, rdict.delta, Ntest, rdict.Ngrid) # smooth PDF generated using input dictionaries
zpdf = zpdf[rdict.zmin_idx_highres:rdict.zmax_idx_highres:rdict.res] / (1+rdict.zgrid) # ln(1+z)->z density
zpdf = interp(rdict.zgrid_out, rdict.zgrid, zpdf)
zpdf /= zpdf.sum() # normalizing
zpdf *= Ntest

# predicted (ver 1)
# this version will be slightly off due to truncations of kernels
zpdf_fz = model_zpdf.sum(axis=0)

# predicted (ver 2)
# this version is exact
zpdf_fz = fz.pdf_kde_dict(rdict.sig_dict, rdict.sig_width, lzidx[train_sel], lzeidx[train_sel], model_pstack,
                          rdict.grid, rdict.delta, Ntrain, rdict.Ngrid) # smooth PDF generated using input dictionaries
zpdf_fz = zpdf_fz[rdict.zmin_idx_highres:rdict.zmax_idx_highres:rdict.res] / (1+rdict.zgrid) # ln(1+z)->z density
zpdf_fz = interp(rdict.zgrid_out, rdict.zgrid, zpdf_fz)
zpdf_fz /= zpdf_fz.sum() # normalizing
zpdf_fz *= Ntest

# plotting
figure(figsize=(14,6))
plot(rdict.zgrid_out, zpdf, lw=5, color='black', label='Underlying')
plot(rdict.zgrid_out, zpdf_fz, lw=5, color='red', alpha=0.7, label='Predicted')
xlim([z.min(), z.max()])
yticks([])
legend(fontsize=20)
tight_layout()
xlabel('Redshift')
ylabel('$N(z|\mathbf{g})$')
```

Our distributions are well within the expected Poisson errors just from random independent sampling, and even the AD metric (which is quite sensitive to differences in the corresponding CDFs) doesn't suggest our distributions are all that different.
