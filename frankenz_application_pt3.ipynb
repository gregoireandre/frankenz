# Application of FRANKEN-Z to Real Data

This notebook demonstrates an application of FRANKEN-Z to S16A **Hyper Suprime-Cam (HSC) Subaru Strategic Program (SSP) survey data**. As the training data includes $\sim\!400,000$ objects that have been **inhomogeneously sampled** (cobbled together from various surveys) with **varying errors** (stochastic backgrounds) and **depths** (collected from wide, deep, and ultra-deep fields) along with **multiple flux measurements** (PSF, Cmodel, and fiber/aperture quantities) and **randomly censored data** (from failed extractions and missing coverage), modeling these data provides many new challenges and opportunities.

This third notebook outlines an application to an unknown dataset. Due to memory constraints on the laptop I'm using, at several points throughout the notebook I restart the kernel.

# Fitting Data

Let's initialize our fluxes, errors, and masks along with our WINBET instances.

```python
# redefining fluxes

p1,v1,m1=flux_afterburner,square(err_afterburner),mask_afterburner # training
p1[m1==False],v1[m1==False]=1.,1. # filling in missing values with arbitrary values
p2,v2,m2=flux_afterburner_test,square(err_afterburner_test),mask_afterburner_test # testing (observed)
p2[m2==False],v2[m2==False]=1.,1. # filling in missing values with arbitrary values
e1,e2=sqrt(v1+square(0.01*p1)),sqrt(v2+square(0.01*p2)) # add 1% error floor

mag1,mage1=fz.asinh_mag_map(p1,e1,zeropoint=flux_zeropoint,skynoise=skynoise) # Luptitude mapping
mag2,mage2=fz.asinh_mag_map(p2,e2,zeropoint=flux_zeropoint,skynoise=skynoise) # Luptitude mapping

del flux_afterburner,err_afterburner,mask_afterburner
del flux_afterburner_test,err_afterburner_test,mask_afterburner_test
```

```python
N_members=25 # number of trees/Monte Carlo realizations used for nearest-neighbor search
N_neighbors=10 # number of neighbors selected at each iteration

winbet_train=fz.WINBET(Ntrees=N_members,Nleaf=N_neighbors)
winbet_test=fz.WINBET(Ntrees=N_members,Nleaf=N_neighbors)
if (m1==False).sum()>0:
    winbet_train.train(p1,v1,m1,mag1,mage1,mdict)
else:
    winbet_train=None
if (m2==False).sum()>0:
    winbet_test.train(p2[flux_sel],v2[flux_sel],m2[flux_sel],mag2[flux_sel],mage2[flux_sel],mdict)
else:
    winbet_test=None
```

With this all set, let's fit the data!

```python
frankenz=fz.FRANKENZ(N_members=N_members,n_neighbors=N_neighbors)
model_obj,model_Nobj,model_ll,model_Nbands=frankenz.predict(p1,e1,m1,
                                                            p2[flux_sel],e2[flux_sel],m2[flux_sel],
                                                            impute_train=winbet_train,impute_test=winbet_test)
```

```python
import sys # system commands
import gc # garbage memory collection
from scipy import stats # statistics
from scipy import special # special functions
from astropy.io import fits # reading fits files

# FRANKEN-Z
import frankenz as fz

%pylab
%matplotlib inline

# re-defining plotting defaults
from matplotlib.font_manager import FontProperties
from matplotlib import gridspec
rcParams.update({'xtick.major.pad': '7.0'})
rcParams.update({'xtick.major.size': '7.5'})
rcParams.update({'xtick.major.width': '1.5'})
rcParams.update({'xtick.minor.pad': '7.0'})
rcParams.update({'xtick.minor.size': '3.5'})
rcParams.update({'xtick.minor.width': '1.0'})
rcParams.update({'ytick.major.pad': '7.0'})
rcParams.update({'ytick.major.size': '7.5'})
rcParams.update({'ytick.major.width': '1.5'})
rcParams.update({'ytick.minor.pad': '7.0'})
rcParams.update({'ytick.minor.size': '3.5'})
rcParams.update({'ytick.minor.width': '1.0'})
rcParams.update({'xtick.color': 'k'})
rcParams.update({'ytick.color': 'k'})
rcParams.update({'font.size': 30})
```

```python
# extracting and discretizing redshifts
hdul=fits.open('/Users/Josh/Dropbox/HSC/HSC_photoz/catalogs/hsc_s16a_combined_specz_highq_clean_errsim_train_v1.fits')
data=hdul[1].data
z,ze,zt,zs=data['redshift'],data['redshift_err'],data['redshift_type'],data['redshift_source']
ze[ze<0]=0.
lzidx,lzeidx=rdict.fit(log(1+z),ze/(1+z)) # discretize redshifts

# clearing up memory
del hdul[1].data,data
for hdu in hdul:
    del hdu
del hdul
```

```python
# extracting fits
hdul=fits.open('/Users/Josh/Dropbox/HSC/HSC_photoz/test_16a/hsc_photoz_s16a_test.fits')
results=hdul[1].data
model_sel=results['model_sel']
model_obj=results['model_obj']
model_Nobj=results['model_Nobj']
model_ll=results['model_ll']
model_Nbands=results['model_Nbands']

Nobs=len(model_sel)

# clearing up memory
del hdul[1].data,results
for hdu in hdul:
    del hdu
del hdul
```

Let's now generate our redshifts. 

```python
# generate redshifts
model_llmin=empty(Nobs,dtype='float32') # min(ln-likelihood)
lzpdf=empty((Nobs,rdict.Nz),dtype='float32') # ln(1+z) PDF
zpdf=empty((Nobs,rdict.Nz_out),dtype='float32') # z PDF
model_levidence=empty(Nobs,dtype='float32') # -2ln(evidence)

for i in xrange(Nobs):
    if i%5000==0: sys.stdout.write(str(i)+' ')
    Nm=model_Nobj[i]
    midx,ll=model_obj[i][:Nm],model_ll[i][:Nm]
    model_llmin[i]=ll.min() # minimum value (for scaling)
    w=exp(-0.5*(ll-model_llmin[i])) # transform to scaled likelihood weights
    model_levidence[i]=-2*log(w.sum())+model_llmin[i] # -2ln(Evidence)
    
    pz=fz.pdf_kde_dict(rdict.sig_dict,rdict.sig_width,lzidx[midx],lzeidx[midx],w,
                       rdict.grid,rdict.delta,rdict.Ngrid) # KDE dictionary PDF
    lzpdf[i]=pz[rdict.zmin_idx_highres:rdict.zmax_idx_highres:int(rdict.res)]
    
# compute Ntype and Ptype
model_Ntype=zeros((Nobs,3),dtype='float32')
model_Ptype=zeros((Nobs,3),dtype='float32')

for i in xrange(Nobs):
    if i%5000==0: sys.stdout.write(str(i)+' ')
    Nm=model_Nobj[i]
    midx,ll=model_obj[i][:Nm],model_ll[i][:Nm]
    ztypes=zt[midx]-1
    like=exp(-0.5*(ll-model_llmin[i]))
    like/=like.sum()
    for j in xrange(Nm):
        model_Ntype[i][ztypes[j]]+=1
        model_Ptype[i][ztypes[j]]+=like[j]
```

## Incorporating Selection Effects

To examine the impact of selection effects on the $dN/dz$ and individual $P(z|\mathbf{F}_g)$'s, we impose a series of flux cuts similar to those used by the HSC weak lensing sample for first year science. This involves two cuts in the $i$-band: a $i_{S/N}>10$ and $i_\textrm{mag}<24.5$.

For two flux cuts in the $i$-band over a flux boundaries $F_{\textrm{cut}}$ and $10\hat{\sigma}_{g,i}$, following Speagle et al. (2017) (see also [*frankenz_tests.ipynb*](https://github.com/joshspeagle/frankenz/blob/master/frankenz_tests.ipynb)) we can write the selection probability $\gamma$ for a given galaxy $g$ as

\begin{equation}
P(\gamma|g)= \left( 1-\frac{1}{2}\textrm{erfc}\left[ \left(\hat{F}_{g,i}-F_{cut}\right) / \sqrt{2}\hat{\sigma}_{g,i} \right] \right)\left( 1-\frac{1}{2}\textrm{erfc}\left[ \left(\hat{F}_{g,i}-10\hat{\sigma}_{g,i}\right) / \sqrt{2}\hat{\sigma}_{g,i} \right] \right).
\end{equation}

```python
# extract i-band cmodel fluxes (training data)
hdul=fits.open('/Users/Josh/Dropbox/HSC/HSC_photoz/catalogs/hsc_s16a_combined_specz_highq_clean_errsim_train_v1.fits')
data=hdul[1].data
icmodel,icmodel_err=data['icmodel_flux'],data['icmodel_flux_err']
icmodel*=10**(-0.4*data['a_i'])
icmodel_err*=10**(-0.4*data['a_i'])
icmodel_mask=(icmodel_err>0.)&isfinite(icmodel_err)&(icmodel!=0.)&isfinite(icmodel)
print 'Objects with missing values:',(icmodel_mask==False).sum()

# set missing values to arbitrary defaults
icmodel[icmodel_mask==False]=1.
icmodel_err[icmodel_mask==False]=1.

# compute magnitude mappings
imag_cmodel,imagerr_cmodel=fz.asinh_mag_map(icmodel,icmodel_err,skynoise,zeropoint=flux_zeropoint)
imagidx_cmodel,imageidx_cmodel=mdict.fit(imag_cmodel,imagerr_cmodel)

# clearing up memory
del hdul[1].data,data
for hdu in hdul:
    del hdu
del hdul
```

```python
# mag cut
icut_mag=24.5
fcut=10**(-0.4*(icut_mag-23.9))

# S/N cut
icut_sn=10.

# impose selection function
isel=(icmodel_test>fcut)&(icmodel_test/icmodel_err_test>icut_sn)&icmodel_mask_test
print 'Objects that pass the selection criteria:',isel.sum()
```

```python
# compute p(\gamma|g) terms
psel1=1-0.5*special.erfc((icmodel_test-fcut)/(sqrt(2)*icmodel_err_test)) # from magnitude cut
psel2=1-0.5*special.erfc((icmodel_test-icut_sn*icmodel_err_test)/(sqrt(2)*icmodel_err_test)) # from S/N cut
psel=(psel1*psel2) # combined

figure(figsize=(10,4))
plot(imag_cmodel_test[isel],psel[isel],'.',color='black',markersize=0.5)
xlabel('$i$-band magnitude')
ylabel('$P(\gamma|g)$')
xlim([23,24.7])
```

$P(\gamma|g)$, however, is not exactly what we want since it traces the observed realization we have applied our selection function to. Instead, we want $P(\gamma|h)$, which signifies **the likelihood of selecting the given galaxy $g$ assuming its intrinsic flux distribution is consistent with that of training galaxy $h$**. This can be written as

\begin{equation}
P(\gamma|h) = \int P(g|h)P(\gamma|g) dg,
\end{equation}

where the integral over $dg$ is taken over all possible flux realizations.

In almost all cases, this integral cannot be evaluated analytically. It can instead be approximated numerically using a series of Monte Carlo fluxes drawn from the flux likelihood (i.e. observed flux distribution) $\lbrace\mathbf{\hat{F}}_g^1,\mathbf{\hat{F}}_g^2,\dots,\mathbf{\hat{F}}_g^n\rbrace \sim \mathcal{N}(\mathbf{\hat{F}}_g|\mathbf{\hat{F}}_{h},\mathbf{\hat{C}}_{h}+\mathbf{\hat{C}}_{g})$ via

\begin{equation}
P(\gamma|h) \approx \frac{1}{n} \sum_{t=1}^{n} P(\gamma|\mathbf{\hat{F}}_g^t,\mathbf{\hat{C}}_g).
\end{equation}

```python
# extracting fits
hdul=fits.open('/Users/Josh/Dropbox/HSC/HSC_photoz/test_16a/hsc_photoz_s16a_test.fits')
results=hdul[1].data
model_sel=results['model_sel']
model_obj=results['model_obj']
model_Nobj=results['model_Nobj']
model_ll=results['model_ll']
model_Nbands=results['model_Nbands']

Nobs=len(model_sel)

# clearing up memory
del hdul[1].data,results
for hdu in hdul:
    del hdu
del hdul
```

```python
# compute p(\gamma|g) terms
N_mc=100
psel_int=zeros(model_ll.shape,dtype='float32') # selection probability integrated over the flux likelihood

counter=0
for j in arange(Nobs)[isel[model_sel]]:
    if counter%5000==0: sys.stdout.write(str(counter)+' ')
    counter+=1
    idx=model_sel[j]
    Nm=model_Nobj[j]
    midx=model_obj[j][:Nm]
    midx=midx[icmodel_mask[midx]]
    midx=array([midx for i in xrange(N_mc)])
    fmodel=icmodel[midx]
    emodel=sqrt(square(icmodel_err[midx])+square(0.01*fmodel)+square(icmodel_err_test[idx]))
    ftemp=normal(fmodel,emodel)
    t1=1-0.5*special.erfc((ftemp-fcut)/(sqrt(2)*icmodel_err_test[idx]))
    t2=1-0.5*special.erfc((ftemp-icut_sn*icmodel_err_test[idx])/(sqrt(2)*icmodel_err_test[idx]))
    psel_int[j][:Nm]+=(t1*t2).sum(axis=0)/N_mc
```

Finally, we can compute the joint probability

\begin{equation}
P(g,\gamma|h)=P(\gamma|h)P(g|h)
\end{equation}

that combines our selection effect and our original likelihood (i.e. how likely we are to **observe and select** $g$ given our belief it is consistent with $h$).

To put all this selection function discussion another way, our original likelihood is just comparing the distance (in flux space) between $g$ and $h$, where deviations away from $g$ are penalized in a symmetric fashion. The selection function breaks this symmetry by favoring deviations towards brighter rather than fainter fluxes around the selection boundary. This is doubly true here since we did not included the Cmodel fluxes in our original likelihoods (and thus are adding in new information).

```python
objid=arange(Nobs)[isel[model_sel]][8] # object id
midx=model_obj[objid][:model_Nobj[objid]] # corresponding training objects

# likelihoods
t1=exp(-0.5*(model_ll[objid][:model_Nobj[objid]]-model_llmin[objid])) 
t1/=t1.sum()

# selection probability
tp=psel_int[objid][:model_Nobj[objid]]
tp/=tp.sum()

# modified probability
t2=t1*tp
t2/=t2.sum()

# compute S/N
ptemp=icmodel[midx]
etemp=sqrt(square(icmodel_err[midx])+square(0.01*ptemp)+square(icmodel_err_test[model_sel[objid]]))
sntemp=ptemp/etemp

# plotting
figure(figsize=(12,9))
subplot(3,2,1)
plot(imag_cmodel[midx],t1,'o',markerfacecolor='blue',markeredgecolor='black')
xlabel('$i$-band magnitude')
locator_params(nbins=3)
ylabel('$P(h|g)$')
yticks([])
tight_layout()
subplot(3,2,2)
plot(sntemp,t1,'o',markerfacecolor='blue',markeredgecolor='black')
xlabel('$i$-band S/N')
locator_params(nbins=3)
ylabel('$P(h|g)$')
yticks([])
tight_layout()
subplot(3,2,3)
plot(imag_cmodel[midx],tp,'o',markerfacecolor='red',markeredgecolor='black')
xlabel('$i$-band magnitude')
locator_params(nbins=3)
ylabel('$P(\gamma|h)$')
yticks([])
tight_layout()
subplot(3,2,4)
plot(sntemp,tp,'o',markerfacecolor='red',markeredgecolor='black')
xlabel('$i$-band S/N')
locator_params(nbins=3)
ylabel('$P(\gamma|h)$')
yticks([])
tight_layout()
subplot(3,2,5)
plot(imag_cmodel[midx],t2,'o',markerfacecolor='darkviolet',markeredgecolor='black')
xlabel('$i$-band magnitude')
locator_params(nbins=3)
ylabel('$P(g,\gamma|h)$')
yticks([])
tight_layout()
subplot(3,2,6)
plot(sntemp,t2,'o',markerfacecolor='darkviolet',markeredgecolor='black')
xlabel('$i$-band S/N')
locator_params(nbins=3)
ylabel('$P(g,\gamma|h)$')
yticks([])
tight_layout()
```

This can substantially modify our original photo-z PDFs for objects near the selection boundary, as shown below.

```python
figure(figsize=(12,4))
pz=fz.pdf_kde_dict(rdict.sig_dict,rdict.sig_width,lzidx[midx],lzeidx[midx],t1,
                   rdict.grid,rdict.delta,rdict.Ngrid) # KDE dictionary PDF
plot(exp(rdict.lzgrid_highres)-1,pz,lw=4,color='black',label='Likelihood only',alpha=0.8)
pz=fz.pdf_kde_dict(rdict.sig_dict,rdict.sig_width,lzidx[midx],lzeidx[midx],tp,
                   rdict.grid,rdict.delta,rdict.Ngrid) # KDE dictionary PDF
plot(exp(rdict.lzgrid_highres)-1,pz,lw=4,color='red',label='Selection only',alpha=0.6)
pz=fz.pdf_kde_dict(rdict.sig_dict,rdict.sig_width,lzidx[midx],lzeidx[midx],t2,
                   rdict.grid,rdict.delta,rdict.Ngrid) # KDE dictionary PDF
plot(exp(rdict.lzgrid_highres)-1,pz,lw=4,color='darkviolet',label='Joint distribution',alpha=0.6)
xlim([0,6])
xlabel('Redshift')
ylabel('$P(z|\mathbf{F}_g)$')
yticks([])
legend(fontsize=16)
```

## Hierchical Modification

We now seek to derive errors on the $dN/dz$ distribution (and the individual PDFs) using our previous hierarchical models. Let's set our population prior $P(\mathbf{f^h})$ and initialize our population parameter vector $\mathbf{f}^{\mathbf{h}}_0$.

```python
p_fh=ones(Ntrain,dtype='float32')/Nobs # uniform prior
```

```python
fh=zeros(Ntrain,dtype='float32')

for i in tsel:
    Nm=model_Nobj[i] # number of models
    midx,ll,mpsel=model_obj[i][:Nm],model_ll[i][:Nm],psel_int[i][:Nm] # model indices, log-likelihoods, and selection effects
    weights=exp(-0.5*(ll-model_llmin[i]))*mpsel # weights
    weights[isfinite(weights)==False]=0 # assign failed computations (nan,inf) 0 weight
    weights*=p_fh[midx] # multiply by global prior
    weights/=weights.sum() # normalize
    fh[midx]+=weights # stack
```

Now that everything's all set up, we generate a bunch of $\mathbf{f^h}$ samples.

```python
N_samples=100
fh_t=zeros((N_samples,len(fh)),dtype='float32')

# Gibbs sampler
fh_t[0]=fh # initialize

for i in xrange(N_samples-1):
    sys.stdout.write(str(i)+' ')
    fh_i=random.dirichlet(fh_t[i]+1) # draw population
    for j in tsel:
        Nm=model_Nobj[j] # number of models
        midx,ll,llmin,mpsel=model_obj[j][:Nm],model_ll[j][:Nm],model_llmin[j],psel_int[j][:Nm]
        weights=exp(-0.5*(ll-llmin))*mpsel
        weights[isfinite(weights)==False]=0 # assign failed computations (nan,inf) 0 weight
        weights*=fh_i[midx]*p_fh[midx] # (P(pop_global)*P(pop_conditional)*likelihood)
        fh_t[i+1][midx]+=weights/weights.sum()
```

As with standard Metropolis-Hastings Markov Chain Monte Carlo (MCMC) sampling, we first have to check if/when our sampler has converged to the stationary distribution.

```python
figure(figsize=(8,4))
plot([correlate(fh_t[i+1],fh_t[i]) for i in xrange(N_samples-1)],color='blue',lw=4)
xlabel('Iteration')
ylabel('Auto-Correlation',fontsize=22)
yticks([])
```

Starting at 40 appears pretty safe.

**Important sidenote**: To properly condition a covariance/precision matrix, we need as many samples as "free parameters". This means we're limited by the resolution of our grid, and that saving on computation time using fewer samples will require downsampling our redshift, mass, etc. grid to coarser resolution using, e.g., binning.

```python
N_burnin=40 # burn-in trials
z_pdf_draws=empty((N_samples-N_burnin,rdict.Nz_out)) # dn/dz draws

for i in xrange(N_burnin,N_samples):
    sys.stdout.write(str(i)+': ')
    midx=choice(Ntrain,p=fh_t[i]/fh_t[i].sum(),size=Nsel) # sample from f^h
    temp1,temp2=lzidx[midx],lzeidx[midx]
    pz=zeros(rdict.lzgrid_highres.shape)
    for j in xrange(Nsel):
        if j%50000==0: sys.stdout.write(str(j/50000)+' ')
        t1,t2=temp1[j],temp2[j]
        w,k=rdict.sig_width[t2],rdict.sig_dict[t2]
        pz[choice(xrange(t1-w,t1+w+1),p=k/k.sum())]+=1
    pz=pz[rdict.zmin_idx_highres:rdict.zmax_idx_highres:int(rdict.res)]
    pz=interp(rdict.zgrid_out,rdict.zgrid,pz/rdict.znorm)
    pz/=pz.sum()
    z_pdf_draws[i-N_burnin]=pz
z_pdf_draws_mean=z_pdf_draws.mean(axis=0)
z_pdf_draws_mean/=z_pdf_draws_mean.sum()
z_pdf_draws_std=z_pdf_draws.std(axis=0)
```

Let's now compare the difference between the original $dN/dz$ and the hierarchical Gibbs draws.

```python
figure(figsize=(14,6))

[plot(rdict.zgrid_out,z_pdf_draws[i],color='blue',lw=3,alpha=0.05) for i in xrange(N_samples-N_burnin)]
plot(rdict.zgrid_out,nz2,lw=5,color='black',label='Original',alpha=0.75)
plot(rdict.zgrid_out,z_pdf_draws[i]*0-0.1,color='blue',lw=3,alpha=0.5,label='Hierarchical',)

xlim([rdict.zgrid_out.min(),rdict.zgrid_out.max()])
legend(fontsize=20)
tight_layout()
xlabel('Redshift')
ylim([0,z_pdf_draws.max()*1.05])
yticks([])
ylabel('dN/dz')
```

That looks pretty good on the whole. On final check is to see how this distribution changes under different prior assumptions. We decide to upweight spec-z's and g/prism-z's relative to photo-z's by a factor of 10 and 5, respectively.

```python
p_fh=ones(Ntrain,dtype='float32')/Nobs # uniform prior
p_fh[zt==1]*=10 # upweight spec-z's by a factor of 10
p_fh[zt==2]*=5 # upweight grism-z's by a factor of 5
p_fh/=p_fh.sum()
```

```python
fh=zeros(Ntrain,dtype='float32')
for i in tsel:
    Nm=model_Nobj[i] # number of models
    midx,ll,mpsel=model_obj[i][:Nm],model_ll[i][:Nm],psel_int[i][:Nm] # model indices, log-likelihoods, and selection effects
    weights=exp(-0.5*(ll-model_llmin[i]))*mpsel # weights
    weights[isfinite(weights)==False]=0 # assign failed computations (nan,inf) 0 weight
    weights*=p_fh[midx] # multiply by global prior
    weights/=weights.sum() # normalize
    fh[midx]+=weights # stack
```

```python
N_samples=100
fh_t=zeros((N_samples,len(fh)),dtype='float32')

# Gibbs sampler
fh_t[0]=fh # initialize

for i in xrange(N_samples-1):
    sys.stdout.write(str(i)+' ')
    fh_i=random.dirichlet(fh_t[i]+1) # draw population
    for j in tsel:
        Nm=model_Nobj[j] # number of models
        midx,ll,llmin,mpsel=model_obj[j][:Nm],model_ll[j][:Nm],model_llmin[j],psel_int[j][:Nm]
        weights=exp(-0.5*(ll-llmin))*mpsel
        weights[isfinite(weights)==False]=0 # assign failed computations (nan,inf) 0 weight
        weights*=fh_i[midx]*p_fh[midx] # (P(pop_global)*P(pop_conditional)*likelihood)
        fh_t[i+1][midx]+=weights/weights.sum()
```

```python
figure(figsize=(8,4))
plot([correlate(fh_t[i+1],fh_t[i]) for i in xrange(N_samples-1)],color='blue',lw=4)
xlabel('Iteration')
ylabel('Auto-Correlation',fontsize=22)
yticks([])
```

```python
N_burnin=20 # burn-in trials
z_pdf_draws=empty((N_samples-N_burnin,rdict.Nz_out)) # dn/dz draws

for i in xrange(N_burnin,N_samples):
    sys.stdout.write(str(i)+': ')
    midx=choice(Ntrain,p=fh_t[i]/fh_t[i].sum(),size=Nsel) # sample from f^h
    temp1,temp2=lzidx[midx],lzeidx[midx]
    pz=zeros(rdict.lzgrid_highres.shape)
    for j in xrange(Nsel):
        if j%50000==0: sys.stdout.write(str(j/50000)+' ')
        t1,t2=temp1[j],temp2[j]
        w,k=rdict.sig_width[t2],rdict.sig_dict[t2]
        pz[choice(xrange(t1-w,t1+w+1),p=k/k.sum())]+=1
    pz=pz[rdict.zmin_idx_highres:rdict.zmax_idx_highres:int(rdict.res)]
    pz=interp(rdict.zgrid_out,rdict.zgrid,pz/rdict.znorm)
    pz/=pz.sum()
    z_pdf_draws[i-N_burnin]=pz
z_pdf_draws_mean=z_pdf_draws.mean(axis=0)
z_pdf_draws_mean/=z_pdf_draws_mean.sum()
z_pdf_draws_std=z_pdf_draws.std(axis=0)
```

```python
figure(figsize=(14,6))

[plot(rdict.zgrid_out,z_pdf_draws[i],color='blue',lw=3,alpha=0.05) for i in xrange(N_samples-N_burnin)]
plot(rdict.zgrid_out,nz2,lw=5,color='black',label='Original',alpha=0.75)
plot(rdict.zgrid_out,z_pdf_draws[i]*0-0.1,color='blue',lw=3,alpha=0.5,label='Hierarchical',)

xlim([rdict.zgrid_out.min(),rdict.zgrid_out.max()])
legend(fontsize=20)
tight_layout()
xlabel('Redshift')
ylim([0,z_pdf_draws.max()*1.05])
yticks([])
ylabel('dN/dz')
```

While there are some small concerning features around $z=2-3$, on the whole this distribution looks quite reasonable.
