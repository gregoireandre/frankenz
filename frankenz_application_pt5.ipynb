# Application of FRANKEN-Z to Real Data

This notebook demonstrates an application of FRANKEN-Z to S16A **Hyper Suprime-Cam (HSC) Subaru Strategic Program (SSP) survey data**. As the training data includes $\sim\!400,000$ objects that have been **inhomogeneously sampled** (cobbled together from various surveys) with **varying errors** (stochastic backgrounds) and **depths** (collected from wide, deep, and ultra-deep fields) along with **multiple flux measurements** (PSF, Cmodel, and fiber/aperture quantities) and **randomly censored data** (from failed extractions and missing coverage), modeling these data provides many new challenges and opportunities.

# Fitting Data

Let's initialize our fluxes, errors, and masks along with our WINBET instances.

```python
# redefining fluxes

p1,v1,m1=flux_cmodel,square(err_cmodel),mask_cmodel # training
p1[m1==False],v1[m1==False]=1.,1. # filling in missing values with arbitrary values
p2,v2,m2=flux_cmodel_test,square(err_cmodel_test),mask_cmodel_test # testing (observed)
p2[m2==False],v2[m2==False]=1.,1. # filling in missing values with arbitrary values
e1,e2=sqrt(v1+square(0.01*p1)),sqrt(v2+square(0.01*p2)) # add 1% error floor

mag1,mage1=fz.asinh_mag_map(p1,e1,zeropoints=flux_zeropoint,skynoise=skynoise) # Luptitude mapping
mag2,mage2=fz.asinh_mag_map(p2,e2,zeropoints=flux_zeropoint,skynoise=skynoise) # Luptitude mapping

del flux_cmodel,err_cmodel,mask_cmodel
del flux_cmodel_test,err_cmodel_test,mask_cmodel_test
```

With this all set, let's fit the data!

```python
N_members, N_neighbors = 25, 10
frankenz=fz.FRANKENZ(N_members=N_members,n_neighbors=N_neighbors)
model_obj,model_Nobj,model_ll,model_Nbands=frankenz.predict(p1,e1,m1,
                                                            p2[flux_sel],e2[flux_sel],m2[flux_sel])
```

## Generating Redshift PDFs

```python
# generate redshifts
model_llmin=empty(Nobs,dtype='float32') # min(ln-likelihood)
lzpdf=empty((Nobs,rdict.Nz),dtype='float32') # ln(1+z) PDF
zpdf=empty((Nobs,rdict.Nz_out),dtype='float32') # z PDF
model_levidence=empty(Nobs,dtype='float32') # -2ln(evidence)

for i in xrange(Nobs):
    if i%5000==0: sys.stdout.write(str(i)+' ')
    Nm=model_Nobj[i]
    midx,ll=model_obj[i][:Nm],model_ll[i][:Nm]
    model_llmin[i]=ll.min() # minimum value (for scaling)
    w=exp(-0.5*(ll-model_llmin[i])) # transform to scaled likelihood weights
    model_levidence[i]=-2*log(w.sum())+model_llmin[i] # -2ln(Evidence)
    
    pz=fz.pdf_kde_dict(rdict.sig_dict,rdict.sig_width,lzidx[midx],lzeidx[midx],w,
                       rdict.grid,rdict.delta,Nm,rdict.Ngrid) # KDE dictionary PDF
    lzpdf[i]=pz[rdict.zmin_idx_highres:rdict.zmax_idx_highres:int(rdict.res)]
    
# compute Ntype and Ptype
model_Ntype=zeros((Nobs,3),dtype='float32')
model_Ptype=zeros((Nobs,3),dtype='float32')

for i in xrange(Nobs):
    if i%5000==0: sys.stdout.write(str(i)+' ')
    Nm=model_Nobj[i]
    midx,ll=model_obj[i][:Nm],model_ll[i][:Nm]
    ztypes=zt[midx]-1
    like=exp(-0.5*(ll-model_llmin[i]))
    like/=like.sum()
    for j in xrange(Nm):
        model_Ntype[i][ztypes[j]]+=1
        model_Ptype[i][ztypes[j]]+=like[j]
```

```python
model_sel=arange(len(p2))[flux_sel]
zpdf=fz.pdfs_resample(rdict.zgrid,lzpdf/(1+rdict.zgrid),rdict.zgrid_out) # resample from ln(1+z) to z space
del lzpdf
```

```python
# color-color cut
"""
Created on Thu Apr 14 16:38:12 2016
Updated Dec 7 2016
@author: elinor
"""

def SelectSources(z,r,i,g, CCparams=[-0.0248, 1.604, 2.276, -0.152], 
                  rlim=[0.5, 21, 28, -0.5, 2.], blim=[-3, -0.8, 22, 28, 0.5, 0.5]):
    """
    SelectSources(table, r,g1,g2,b, CCparams, rlim, blim)
    
    """
    aCM, bCM, aCC, bCC = CCparams
    Z = z
    RZ = r - z
    GI = g - i
    
    f1 = aCM*Z + bCM # color-mag R-I seq
    seqdif1 = RZ - f1
    CCf = aCC * RZ + bCC          #color-color cluster sequence
    CCf2 = -1./aCC*RZ - bCC/aCC**2 # line perpendicular to color-color sequence
    CCdif = GI-CCf                 # B-R -CC Sequence
    CCdif2 = (GI-CCf2)/(1.+1./aCC**2)# B-R -perCC Sequence

    
    r_rzlim    = ( RZ > rlim[0] ) # RZ lower limit, separate red from blue
    r_maglim   =  ( Z > rlim[1] ) & ( Z < rlim[2] ) # magnitude limit, redderst band
    r_CCseqlim = ( CCdif < rlim[3] ) &  ( CCdif2 < rlim[4] ) 
    red = r_maglim & r_rzlim & r_CCseqlim
    
    b_CMseqlim = (seqdif1 < blim[1]) & (seqdif1>blim[0]) & (GI<4)
    b_maglim   =  ( Z > blim[2] ) & ( Z < blim[3] ) # magnitude limit, redderst band
    b_rzlim    = ( RZ < blim[4]) # RZ upper limit, separate red from blue
    b_CCseqlim = ( CCdif2 < blim[5] )
    blue =  b_rzlim & (b_CCseqlim |  b_CMseqlim )  & b_maglim
    return red, blue
```

```python
from astropy import cosmology
from astropy import units as u
cosmo = cosmology.FlatLambdaCDM(100, 0.3)
def Sigma_cr_co(zl, z):
    """ calculate comoving Sigma_cr using full P(z) integration """
    c = 9.7156119e-15*(u.Mpc/u.s)
    G = 4.5183594e-48*(u.Mpc**3/u.Msun/u.s**2)
    sigma_cr_fact = c**2 / (4*np.pi*G) # physical factors
    chi_zs= cosmo.comoving_distance(z) #source comoving distance 
    chi_zl = cosmo.comoving_distance(zl) # lens comoving ditance
    ad_zl = cosmo.angular_diameter_distance(zl) # lens angular diameter distance 
    if z>zl:
        Sigma_cr = sigma_cr_fact / (1 - chi_zl/chi_zs)*z / ad_zl / (1+zl)**2 # Sigma_cr calculation masking below lens redshift
    else:
        Sigma_cr = inf
    return Sigma_cr
```

```python
zr, zb = [], []
shape_weight = data['ishape_hsm_regauss_derived_weight']
zgrid_hr = exp(rdict.lzgrid_highres)
zlens = 0.5 # fixed lens redshift
lens_weight = array([1. / Sigma_cr_co(zlens, z).value**2 for z in zgrid_hr])

for j in xrange(20):
    sys.stdout.write(str(j)+' ')
    
    # Monte Carlo fluxes
    g,r,i,z,y = fz.asinh_mag_map(normal(p2,e2)[flux_sel],e2[flux_sel],skynoise,zeropoints=flux_zeropoint)[0].swapaxes(0,1)
    
    # select sample
    rsel, bsel = SelectSources(z,r,i,g)

    # red sample
    z_pdf_r = zeros(rdict.Nz_highres)
    for i in arange(Nobs)[rsel]:
        Nm=model_Nobj[i] # number of models
        midx,ll=model_obj[i][:Nm],model_ll[i][:Nm] # model indices, log-likelihoods
        model_llmin[i]=ll.min() # minimum value (for scaling)
        w=float64(exp(-0.5*(ll-model_llmin[i]))) # transform to scaled likelihood weights
        m = choice(midx,p=w/w.sum()) # sample object
        zt, zet = lzidx[m], lzeidx[m] # grab discretized redshift index and associated redshift kernel index
        zkern, zwkern = rdict.sig_dict[zet], rdict.sig_width[zet] # grab kernel and width from dictionary
        redshift = choice(arange(zt-zwkern,zt+zwkern+1),p=zkern/zkern.sum()) # sample redshift index from kernel
        z_pdf_r[redshift]+=shape_weight[i]*lens_weight[redshift] # add shape weight * lens weight
    z_pdf_r=z_pdf_r[rdict.zmin_idx_highres:rdict.zmax_idx_highres]/(1+rdict.zgrid_highres) # convert from ln(1+z) to z
    z_pdf_r=interp(rdict.zgrid_out,rdict.zgrid_highres,z_pdf_r) # interpolate to uniform grid in z
    z_pdf_r/=z_pdf_r.sum() # normalize
    zr.append(copy(z_pdf_r)) # append
    
    # blue sample
    z_pdf_b = zeros(rdict.Nz_highres)
    for i in arange(Nobs)[bsel]:
        Nm=model_Nobj[i] # number of models
        midx,ll=model_obj[i][:Nm],model_ll[i][:Nm] # model indices, log-likelihoods
        model_llmin[i]=ll.min() # minimum value (for scaling)
        w=float64(exp(-0.5*(ll-model_llmin[i]))) # transform to scaled likelihood weights
        m = choice(midx,p=w/w.sum()) # sample object
        zt, zet = lzidx[m], lzeidx[m] # grab discretized redshift index and associated redshift kernel index
        zkern, zwkern = rdict.sig_dict[zet], rdict.sig_width[zet] # grab kernel and width from dictionary
        redshift = choice(arange(zt-zwkern,zt+zwkern+1),p=zkern/zkern.sum()) # sample redshift index from kernel
        z_pdf_b[redshift]+=shape_weight[i]*lens_weight[redshift] # add shape weight * lens weight
    z_pdf_b=z_pdf_b[rdict.zmin_idx_highres:rdict.zmax_idx_highres]/(1+rdict.zgrid_highres) # convert from ln(1+z) to z
    z_pdf_b=interp(rdict.zgrid_out,rdict.zgrid_highres,z_pdf_b) # interpolate to uniform grid in z
    z_pdf_b/=z_pdf_b.sum() # normalize
    zb.append(copy(z_pdf_b)) # append
```
